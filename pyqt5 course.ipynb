```python
from PyQt5 import QtWidgets as qtw
from PyQt5 import QtGui as qtg
from PyQt5 import QtCore as qtc
import sys
import mysql.connector


class MainWindow(qtw.QWidget):
    def __init__(self):
        super().__init__()
        #### left layout ####
        self.mydb=mysql.connector.connect(host='localhost',user='root',password='2182003s2gbs')
        self.my_cursor=self.mydb.cursor()
        try:
            self.my_cursor.execute("CREATE DATABASE students")
            self.mydb=mysql.connector.connect(host='localhost',user='root',password='2182003s2gbs',database="students")
            self.my_cursor=self.mydb.cursor()
            
        except:
            self.mydb=mysql.connector.connect(host='localhost',user='root',password='2182003s2gbs',database="students")
            self.my_cursor=self.mydb.cursor()
        try:
            self.my_cursor.execute("CREATE TABLE students_data (name VARCHAR(100) PRIMARY KEY ,age INTEGER(10))")
        except:
            pass
            
        self.cmb=qtw.QComboBox()
        self.cmb.addItems(["add students","search student","open image"])
        self.cmb.currentIndexChanged.connect(self.select_layout)
        
        #### First layout ingredients ####
        self.firstName=qtw.QLineEdit()
        self.LastName=qtw.QLineEdit()
        self.firstName.setPlaceholderText("enter your first name")
        self.LastName.setPlaceholderText("enter your last name")
        self.btn=qtw.QPushButton("")
        self.btn.setFlat(True)
        self.btn.clicked.connect(self.submit_data)
        pix=qtg.QPixmap("img_icon")#any icon
        self.btn.setIcon(qtg.QIcon(pix))
        self.sb=qtw.QSpinBox()
        self.sb.setRange(0,100)
        self.sb.setWrapping(True)
        self.sb.setSingleStep(1)
        
        #### second layout ingredients ####
        self.search=qtw.QLineEdit()
        self.search.setPlaceholderText("Search")
        self.search.editingFinished.connect(self.get_data)
        self.label=qtw.QLabel("Label")
        
        
        #### third layout ingredients ####
        self.open_img_btn=qtw.QPushButton("open image")
        self.open_img_btn.clicked.connect(self.file_dialog)
        self.img_lbl=qtw.QLabel("Image label")
        self.setupLayout()
        self.show()
    
    def setupLayout(self):
        self.mainlayout=qtw.QHBoxLayout()
        self.leftLayout=qtw.QVBoxLayout()
        self.SLayout=qtw.QStackedLayout()
        
        #### left layout ####
        self.leftLayout.addWidget(self.cmb)
        #### stacked layout ####
        
        ## add student ##
        self.addStudWid=qtw.QWidget()
        self.addStudLayout=qtw.QFormLayout()
        self.addStudLayout.addRow("First Name: ",self.firstName)
        self.addStudLayout.addRow("Last Name: ",self.LastName)
        self.addStudLayout.addRow("Age:  ",self.sb)
        self.addStudLayout.addRow("",self.btn)
        self.addStudWid.setLayout(self.addStudLayout)
        ## search Student ##
        self.searchStudWid=qtw.QWidget()
        self.searchStudlayout=qtw.QVBoxLayout()
        self.searchStudlayout.addWidget(self.search)
        self.searchStudlayout.addStretch()
        self.searchStudlayout.addWidget(self.label)
        self.searchStudWid.setLayout(self.searchStudlayout)
        ## third Layout ##
        self.openImgWid=qtw.QWidget()
        self.openImglayout=qtw.QVBoxLayout()
        self.openImglayout.addWidget(self.img_lbl)
        self.openImglayout.addStretch()
        self.openImglayout.addWidget(self.open_img_btn)
        self.openImgWid.setLayout(self.openImglayout)
        
        ## stacked layout its self ##
        self.SLayout.addWidget(self.addStudWid)
        self.SLayout.addWidget(self.searchStudWid)
        self.SLayout.addWidget(self.openImgWid)
        self.SWidget=qtw.QWidget()
        self.SWidget.setLayout(self.SLayout)
        ## left layout##
        self.leftWid=qtw.QWidget()
        self.leftWid.setLayout(self.leftLayout)
        ##main layout ##
        self.mainlayout.addWidget(self.leftWid)
        self.mainlayout.addStretch()
        self.mainlayout.addWidget(self.SWidget)
        self.setLayout(self.mainlayout)
        
    def select_layout(self,idx):
        self.SLayout.setCurrentIndex(idx)
        self.cmb.adjustSize()
        
    def file_dialog(self):
        res=qtw.QFileDialog.getOpenFileName(self,'Open File','C:\\Users\\abdel\\OneDrive\\Desktop',"Image files (*.jpg *.gif *.png *.jpeg)")
        pixmap=qtg.QPixmap(res[0])
        pixmap=pixmap.scaledToWidth(300)
        pixmap=pixmap.scaledToHeight(300)
        self.img_lbl.setPixmap(pixmap)
        self.img_lbl.adjustSize()
    def submit_data(self):
        name=self.firstName.text()+" "+self.LastName.text()
        age=int(self.sb.value())
        try:
            sql_stuff="INSERT INTO students_data (name,age) VALUES (%s,%s)"
            self.my_cursor.execute(sql_stuff,(name,age))
            self.mydb.commit()
        except:
            sql_stuff="UPDATE students_data SET age=%s WHERE name=%s"
            self.my_cursor.execute(sql_stuff,(age,name))
            self.mydb.commit()
            
        self.firstName.setText("")
        self.LastName.setText("")
        self.sb.setValue(0)
    def get_data(self):
        self.my_cursor.execute("SELECT * FROM students_data")
        data=self.my_cursor.fetchall()
        name=self.search.text()
        for nam,age in data:
            if nam.lower()==name.lower():
                self.label.setText(f"Name:{name} \n Age:{age}")
                self.label.adjustSize()
                break
        
        
        
if __name__=='__main__':
    app=qtw.QApplication([])
    mw=MainWindow()
    app.exec_()
      

```

```python
import PyQt5.QtWidgets as qtw
import PyQt5.QtGui as qtg
import PyQt5.QtCore as qtc
import PyQt5.QtWebEngineWidgets as qtwe
import glob
import sys
import random

class MainWindow(qtw.QWidget):
    def __init__(self):
        super().__init__()
        self.classes=[]
        self.wev=qtwe.QWebEngineView()
        self.wev.setUrl(qtc.QUrl.fromUserInput("https://doc.qt.io/qt-5/qwidget.html"))
        self.searchLabel=qtw.QLabel()
        self.searchLabel.setPixmap(qtg.QPixmap("search.png").scaledToWidth(20).scaledToHeight(20))
        self.backbutton=qtw.QPushButton("")
        self.backbutton.setIcon(qtg.QIcon(qtg.QPixmap("arrow-180.png")))
        self.forwardbutton=qtw.QPushButton("")
        self.forwardbutton.setIcon(qtg.QIcon(qtg.QPixmap("arrow.png")))
        self.searchBox=qtw.QLineEdit()
        self.searchBox.editingFinished.connect(self.search_class)
        self.home=qtw.QPushButton("")
        self.home.setIcon(qtg.QIcon(qtg.QPixmap("home-big.png")))
        self.home.clicked.connect(self.take_home)
        self.history=qtw.QPushButton("")
        self.history.setIcon(qtg.QIcon(qtg.QPixmap("history.png")))
        self.history.clicked.connect(self.get_history)
        
        self.backbutton.clicked.connect(self.back)
        self.forwardbutton.clicked.connect(self.forward)

        self.tree=qtw.QTreeWidget()
        self.populate_tree()
        self.tree.sortItems(0,qtc.Qt.AscendingOrder)
        self.tree.itemDoubleClicked.connect(self.show_class)
        self.iterate_childs()
        self.classes=sorted(self.classes)
        
        self.cmb=qtw.QComboBox()
        self.cmb.addItems(self.classes)
        self.lineEdit=qtw.QLineEdit()
        self.lineEdit.editingFinished.connect(self.add_item)
        
        self.tabel=qtw.QTableWidget(4,3)
        self.tabel.setColumnWidth(0,100)
        self.tabel.setColumnWidth(2,450)
        self.tabel.setColumnWidth(0,450)

        self.tabel.hide()
        
        self.tabel.setHorizontalHeaderLabels(["Name","URL","Time"])
        self.tabel.itemDoubleClicked.connect(self.take_me_there)
        
        
        self.iterate_childs()
        self.setupLayout()
        
    def take_me_there(self):
        row=self.tabel.currentRow()
        it=self.tabel.item(row,1).text()
        self.wev.setUrl(qtc.QUrl.fromUserInput(it))
        self.wev.repaint()
    def populate_tabel(self,myList):
        self.tabel.setRowCount(len(myList))
        for ind,i in enumerate(myList):
            title,url,time=i
            self.tabel.setItem(ind,0,qtw.QTableWidgetItem(str(title)))
            self.tabel.setItem(ind,1,qtw.QTableWidgetItem(str(url)))
            self.tabel.setItem(ind,2,qtw.QTableWidgetItem(str(time)))

    def get_history(self):
        if self.tabel.isHidden():
            myList=[]
            for i in self.wev.history().items():
                #print(i.title(),i.lastVisited(),i.url())
                title=i.title().split('|')[0].split()[0]
                time=i.lastVisited()
                time=time.currentDateTime().toString()
                url=i.url().toString()
                myList.append([title,url,time])
            self.populate_tabel(myList)
            self.tabel.show()
        else:
            self.tabel.hide()
    def take_home(self):
        self.wev.setUrl(qtc.QUrl.fromUserInput("https://doc.qt.io/qt-5/qwidget.html"))
        self.wev.repaint()
    def add_item(self):
        ans=qtw.QMessageBox.question(self,"classes",f"do yo realy want to add {self.lineEdit.text()} class under the {self.cmb.currentText()} class")
        if ans==qtw.QMessageBox.Yes:
            twi=self.tree.findItems(self.cmb.currentText(),qtc.Qt.MatchRecursive)[0]
            twi.addChild(qtw.QTreeWidgetItem([self.lineEdit.text(),str(random.randrange(25)),str(random.randrange(5))]))
            self.classes.append(self.lineEdit.text())
            self.classes=sorted(self.classes)
            self.cmb.addItem(self.lineEdit.text())
            self.repaint()
        else:
            pass
        
    
    def back(self):
        self.wev.back()
        self.wev.repaint()
    def forward(self):
        self.wev.forward()
        self.wev.repaint()
    def search_class(self):
        itm=self.searchBox.text()
        self.wev.setUrl(qtc.QUrl(f"https://doc.qt.io/qt-5/{itm.lower()}.html"))
        self.wev.repaint()
    def populate_tree(self):
        self.tree.setColumnCount(3)
        self.tree.setHeaderLabels(["Class","Methods","Signals"])
        self.QW=qtw.QTreeWidgetItem(self.tree,["QWidget Module","10","-"])
        self.QW.setToolTip(0,"shows sub classes of QWidget")#col 0,string

        self.QG=qtw.QTreeWidgetItem(self.tree,["QGui Module","10","-"])
        self.QC=qtw.QTreeWidgetItem(self.tree,["QCore Module","10","-"])
        ### add sub-children
        QWSubClasses=["QPushButton","QLabel","QDialog","QTabWidget","QListWidget","QLineEdit","QCheckBox","QGroupBox","QDateEdit","QTimeEdit","QDateTimeEdit","QRadioButton","QTextEdit","QTreeWidget","QSpinBox","QComboBox"]
        for cls in QWSubClasses:
            self.QW.addChild(qtw.QTreeWidgetItem([cls,str(random.randrange(25)),str(random.randrange(5))]))
        QGSubClasses=["QFont","QImage","QPixmap","QIcon","QColor"]
        for cls in QGSubClasses:
            self.QG.addChild(qtw.QTreeWidgetItem([cls,str(random.randrange(25)),str(random.randrange(5))]))
        QCSubClasses=["QDate","QDateTime","QTime","QThread","QFile"]
        for cls in QCSubClasses:
            self.QC.addChild(qtw.QTreeWidgetItem([cls,str(random.randrange(25)),str(random.randrange(5))]))
        ### add sub classes for QDialog
        QDialogTreeWidgetItem=self.tree.findItems("QDialog",qtc.Qt.MatchRecursive)[0]#because it returns a list,MatchRecursive searches all levels 
        QDSubClasses=["QFileDialog","QInputDialog","QMessageBox","QColorDialog","QFontDialog"]
        for cls in QDSubClasses:
            QDialogTreeWidgetItem.addChild(qtw.QTreeWidgetItem([cls,str(random.randrange(25)),str(random.randrange(5))]))
        

    def show_class(self,itm,col):
        #qtw.QMessageBox.information(self,"Class name",f"you chose {itm.text(0)} class")
        self.wev.setUrl(qtc.QUrl(f"https://doc.qt.io/qt-5/{itm.text(0).lower()}.html"))
        self.wev.repaint()
    def get_sub_childs(self,twi):
        self.classes.append(twi.text(0))
        if(twi.childCount()>0):
            for i in range(twi.childCount()):
                self.get_sub_childs(twi.child(i))
                
    def iterate_childs(self):
        for i in range(self.tree.topLevelItemCount()):
            self.get_sub_childs(self.tree.topLevelItem(i))
        
    
        
        
        
    def setupLayout(self):
        self.mainLayout=qtw.QHBoxLayout()
        self.engineLayout=qtw.QVBoxLayout()
        self.treeLayout=qtw.QVBoxLayout()
        self.search_barLayout=qtw.QHBoxLayout()
        self.search_barLayout.addWidget(self.backbutton)
        self.search_barLayout.addWidget(self.forwardbutton)
        self.search_barLayout.addWidget(self.home)
        self.search_barLayout.addWidget(self.history)
        #self.search_barLayout.addStretch()
        self.search_barLayout.addWidget(self.searchLabel)
        self.search_barLayout.addWidget(self.searchBox)
        self.search_barLayout.addStretch()
        self.engineLayout.addLayout(self.search_barLayout)
        self.engineLayout.addWidget(self.wev)
        self.engineLayout.addWidget(self.tabel)
        self.engineLayout.addStretch()
        
        
        self.treeLayout.addWidget(self.tree)
        self.treeLayout.addWidget(self.cmb)
        self.treeLayout.addWidget(self.lineEdit)
        self.mainLayout.addLayout(self.treeLayout)
        self.mainLayout.addLayout(self.engineLayout)

        self.setLayout(self.mainLayout)
        self.show()
        
        
if __name__=="__main__":
    app=qtw.QApplication([])
    mw=MainWindow()
    app.exec_()
```

```python
import PyQt5.QtWidgets as qtw
import PyQt5.QtGui as qtg
import PyQt5.QtCore as qtc

class MainWindow(qtw.QWidget):##Qgroup box is a container of items with sth like a frame around them that seprates them from the rest of the window
    # and can conatin other widgets   not just radio buttons
    #QGRoup boxes can have layouts too
    def __init__(self):
        super().__init__()
        self.setWindowTitle("QGroup Box")
        
        
        self.groupbox_1=qtw.QGroupBox("flat")
        self.rbtTrueFlat=qtw.QRadioButton("True")
        self.rbtFalseFlat=qtw.QRadioButton("False")
        self.rbtTrueFlat.toggled.connect(self.check_flat)
        
        
        
        self.groupbox_2=qtw.QGroupBox("checkable")
        self.rbtTruecheckable=qtw.QRadioButton("True")
        self.rbtFalsecheckable=qtw.QRadioButton("False")
        self.rbtTruecheckable.toggled.connect(self.check_checkable)
        
        
        
        self.groupbox_3=qtw.QGroupBox("alligment")
        self.rbtleft=qtw.QRadioButton("left")
        self.rbtright=qtw.QRadioButton("right")
        self.rbtcenter=qtw.QRadioButton("center")
        self.rbtleft.toggled.connect(self.leftal)
        self.rbtright.toggled.connect(self.rightal)
        self.rbtcenter.toggled.connect(self.centeral)
        
        
        self.setupLayout()
    def leftal(self,chck):
        if chck:
            self.groupbox_3.setAlignment(qtc.Qt.AlignLeft)
            self.groupbox_3.repaint()
    def rightal(self,chck):
        if chck:
            self.groupbox_3.setAlignment(qtc.Qt.AlignRight)
            self.groupbox_3.repaint()
    def centeral(self,chck):
        if chck:
            self.groupbox_3.setAlignment(qtc.Qt.AlignCenter)
            self.groupbox_3.repaint()
    def check_flat(self,flat):
        self.groupbox_1.setFlat(flat)
    def check_checkable(self,check):
        self.groupbox_2.setCheckable(check)
        
    def setupLayout(self):
        self.MainLayout=qtw.QHBoxLayout()
        self.gp1Layout=qtw.QVBoxLayout()
        self.gp1Layout.addWidget(self.rbtTrueFlat)
        self.gp1Layout.addWidget(self.rbtFalseFlat)

        self.groupbox_1.setLayout(self.gp1Layout)
        self.MainLayout.addStretch()
        self.MainLayout.addWidget(self.groupbox_1)
        
        
        self.gp2Layout=qtw.QVBoxLayout()
        self.gp2Layout.addWidget(self.rbtTruecheckable)
        self.gp2Layout.addWidget(self.rbtFalsecheckable)

        self.groupbox_2.setLayout(self.gp2Layout)
        self.MainLayout.addWidget(self.groupbox_2)
        
        self.gp3Layout=qtw.QVBoxLayout()
        self.gp3Layout.addWidget(self.rbtleft)
        self.gp3Layout.addWidget(self.rbtright)
        self.gp3Layout.addWidget(self.rbtcenter)
        
        self.groupbox_3.setLayout(self.gp3Layout)
        self.MainLayout.addWidget(self.groupbox_3)
        
        self.MainLayout.addStretch()
        
        self.setLayout(self.MainLayout)
        self.show()
        

if __name__=="__main__":
    app=qtw.QApplication([])
    mw=MainWindow()
    app.exec_()
```

## QTimer

```python
import PyQt5.QtWidgets as qtw
import PyQt5.QtCore as qtc
import time as t
class MainWindow(qtw.QWidget):
    def __init__(self):
        super().__init__()
        self.prg=qtw.QProgressBar()
        self.prg.setTextVisible(True)
        self.my_prg=0
        self.timer=qtc.QTimer()
        self.timer.start(10000)
        self.timer.timeout.connect(self.prg_dis)
        self.ly=qtw.QVBoxLayout()
        self.ly.addWidget(self.prg)
        self.setLayout(self.ly)
        self.show()
    def prg_dis(self):
        self.my_prg+=10
        self.prg.setValue(self.my_prg)
        self.timer.setInterval(1000)
        if(self.my_prg==100):
            self.timer.stop()
    
if __name__=="__main__":
    app=qtw.QApplication([])
    mw=MainWindow()
    app.exec_()
```

### QStyleFactory

```python
import PyQt5.QtWidgets as qtw
import PyQt5.QtCore as qtc
import PyQt5.QtGui as qtg
import time as t
class MainWindow(qtw.QWidget):
    def __init__(self):
        super().__init__()
        s=qtw.QStyleFactory().keys()
        #print(s)
        #self.setStyle(qtw.QStyleFactory().create(s))
        self.btn_1=qtw.QPushButton("Button 1")
        self.btn_2=qtw.QPushButton("Button 2")
        self.btn_3=qtw.QPushButton("Button 3")
        self.label_1=qtw.QLabel("Abdelrahman")
        self.label_1.setAlignment(qtc.Qt.AlignCenter)
        self.le=qtw.QLineEdit()
        self.sb=qtw.QSpinBox()
        
        self.list=qtw.QListWidget()
        for i in s:
            self.list.addItem(qtw.QListWidgetItem(i))
        self.list.itemDoubleClicked.connect(self.change)
        self.ly=qtw.QVBoxLayout()
        self.ly.addWidget(self.label_1)
        self.ly.addWidget(self.btn_1)
        self.ly.addWidget(self.btn_2)
        self.ly.addWidget(self.btn_3)
        self.ly.addWidget(self.le)
        self.ly.addWidget(self.sb)
        self.ly.addWidget(self.list)
        self.setLayout(self.ly)
        self.show()
    def change(self,itm):
        ret=qtw.QMessageBox.question(self,"Style",f"do you want to change your style to {itm.text()}")
        if ret==qtw.QMessageBox.Yes:
            self.setStyle(qtw.QStyleFactory().create(itm.text()))
            self.repaint()
        else:
            pass
if __name__=="__main__":
    app=qtw.QApplication([])
    mw=MainWindow()
    app.exec_()
```

## QDarkMode

```python
import qdarkstyle as qd
import PyQt5.QtWidgets as qtw
import PyQt5.QtGui as qtg


class MainWindow(qtw.QWidget):
    def __init__(self):
        super().__init__()
        self.is_dark=False
        self.gpb=qtw.QGroupBox()
        self.normal=qtw.QRadioButton("Normal")
        self.dark=qtw.QRadioButton("dark")
        self.normal.toggled.connect(self.make_dark)
        self.normal.setStyleSheet(self.RadiButtonStyler())
        self.dark.setStyleSheet(self.RadiButtonStyler())

        self.ly=qtw.QVBoxLayout()
        self.ly.addWidget(self.normal)
        self.ly.addWidget(self.dark)
        self.txtBox=qtw.QTextEdit()
        self.ly_2=qtw.QHBoxLayout()
        self.ly_2.addLayout(self.ly)
        self.ly_2.addWidget(self.txtBox)
        self.setLayout(self.ly_2)
        self.show()
    def make_dark(self,chk):
        if not chk:
            app.setStyleSheet(qd.load_stylesheet_pyqt5())
            self.is_dark=False
        else:
            self.is_dark=True
            app.setStyleSheet("")
    def RadiButtonStyler(self):
        if not self.is_dark:
            s="""
            
            
            QRadioButton::indicator {background-color:black;color:white;}
            QRadioButton{
            background-color:black;color:white;
            
            }
            
            """
        else:
            s="""
            
            
            QRadioButton::indicator {background-color:white;color:black;}
            QRadioButton{
            background-color:blue;color:blue;
            
            }
            
            """
        return s
        
        
        
if __name__=="__main__":
    app=qtw.QApplication([])
    mw=MainWindow()
    app.exec_()
```
